# Monte Carlo Analysis of LDPC Decoders

## Imports

```python
import sys
import time
sys.path.insert(0,'..')
```

```python
import numpy as np
import matplotlib.pyplot as plt

from simulators import MonteCarloSimulator
from channels import BinarySymmetricChannel, BinaryAWGNChannel, BinaryErasureChannel
from decoders import Gallager, BeliefPropagation, MinSum, IterativeErasure, Peeling
from encoders import BasicEncoder
from utils import read_ldpc_graph
```

```python
# Read Tanner Graph and specify simulation parameters
G = read_ldpc_graph('../data/Tanner.dat')
encoder = BasicEncoder(G)
numtrials = int(1e3)
```

## Binary Symmetric Channel

```python
# Set up encoder-channel-decoder
alpha = np.linspace(0.01, 0.2, 20)
channel = BinarySymmetricChannel()
dec_name = ['Gallager-B', 'Belief Propagation', 'Min Sum', 'Self Corrected Min Sum']
dec_list = [Gallager(G, variant='B'), BeliefPropagation(G, channel), MinSum(G), MinSum(G, self_correcting=True)]
sim_list = [MonteCarloSimulator(encoder, decoder, channel) for decoder in dec_list]
frame_error_rate = np.zeros((len(sim_list), len(alpha)))
```

```python
# Monte Carlo Analysis
for i, (name, sim) in enumerate(zip(dec_name, sim_list)):
    
    # Start the clock
    start_time = time.time()
    
    # Run Simulation
    for j, a in enumerate(alpha):
        channel.alpha = a
        frame_error_rate[i, j] = sim.run(numtrials=numtrials)

    # Time the Simulation
    exec_time = time.time() - start_time
    print(f'Simulation Completed in {exec_time:.3f} seconds -> {name} Decoder')
```

```python
# Plot Data
for i, name in enumerate(dec_name):
    plt.plot(alpha, frame_error_rate[i, :], linewidth=1.5, label=name)

# Format plot
plt.title('Decoding Performance - BSC')
plt.xlabel('alpha')
plt.ylabel('FER')
plt.grid(True)
plt.legend()
```
